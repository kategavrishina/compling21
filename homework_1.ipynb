```python
import re
```

```python
def remove_tags(text):
    # remove tags
    text = re.sub("</.+?>", "", text) # чтобы закрывающие теги перед знаками препинания не оставляли после себя пробелы (...CrazyEgg</a>,...)
    text = re.sub("<.+?>", " ", text)
    return re.sub(" +", " ", text)
```

Разница очевидно есть.

Особенности mystem:
*   приводит глаголы совершенного вида к несовершенному
*   "предпочитает" наречия (сравнительные формы приводит к форме наречия, а не прилагательного - "больше", "точнее", "теплее")
*   в сложных спорных случаях либо оставляет незнакомое слово, как есть ("врядли", "юзабилити"), либо максимум меняет/опускает окончание ("конфети" (конфеть), "юзабилиста" (юзабилист))
*   лучше справляется с неинтуитивными русскими словами вроде "деньги", "данные" (не наречие, а сущ), "заполненнее"
*   правильно лемматизирует местоимения, различает разные начальные формы (это vs этот, то vs тот)

Особенности pymorphy:
*   использует букву ё!
*   не приводит глаголы совершенного вида к несовершенному
*   иначе справляется с местоимениями (сложно сказать, хуже или лучше, потому что есть спорные случаи - нужно ли приводить "им" к форме "он" (майстем) или "они" (пайморфи)? можно ли все производные местоимения "это" приводить к форме "это", как пайморфи, или лучше различать "этот" и "это", например?)
*   распознал сокращение "т.е."! в отличие от майстема
*   "деньга" :(
*   справился лучше майстема со словом "стоит" (= стоить)
*   иногда приводит странные неначальные формы вроде "больший", "малый", "среднее"

Однозначного ответа, что лучше, у меня нет, но, похоже, что у пайморфи минусов всё-таки больше. Если не знать о преимуществах пайморфи (то есть анализировать без сравнения), майстем кажется отличным вариантом. И нельзя забывать, что здесь выбирался самый вероятный из предложенных разборов, так что, возможно, среди менее вероятных можно было найти правильный.

```python
lemmas_smart = re.findall('(?<={)\w+?(?=[}\?])', parsed)
tokens = re.findall('(?<=\W)\w+?(?={)', parsed)
```

```python
for x, y, z in zip(range(len(tokens)), lemmas_normal, lemmas_smart):
    if y != z:
        print('Токен: {0}\nконтекст: {1}\nлемма: {2}\nлемма с дизамбигуацией: {3}\n\n'.format(tokens[x], " ".join(tokens[x-2:x+5]), y, z))
```

Да, различия есть. Опять же сложно судить, какой способ лучше, но вот некоторые замечания:

Преимущества дизамбигуации:
* "прошедший" вместо "прошедшее" (Новый год)
* "стоить" вместо "стоять" (про стоимость стартапа)
* собственно "стартап" вместо "стартапа"
* "быть" вместо "будет" (странно, что будущее время даётся как начальная форма)
* "этот" вместо "это" (как я поняла, это всё же правильная начальная форма для подобных местоимений)
* "много" вместо "больше" (про кол-во конфетти)
* союз "тем" вместо "то" (по контексту)

Недостатки:
* не преобразует сравнительные формы в начальные
* "тепло" вместо "теплый", хотя по контексту прилагательное
* явно не начальные формы: "разумеется", "они", "может"
* "есть" вместо "быть"

С причастием "говорящий" сложный вопрос, так как иногда считается, что начальная форма причастия - глагол, но мне всё же кажется, что правильнее определять причастие в И.п. муж. рода в ед. числе (как в случае с дизамбигуацией).
